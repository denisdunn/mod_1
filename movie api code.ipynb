```python
import time
```

```python
#created a function that pulls a list of 
def get_popular_movies_TMDB(api_key, pages):
    
    url = 'https://api.themoviedb.org/3/movie/popular?api_key={}&language=en-US&page={}'.format(api_key, pages)
    
    response = requests.get(url)
    
    if response.status_code==200:
        
        return response.json()['results']
    
    else:
        return response.status_code


```

```python
get_popular_movies_TMDB(api_key, 21)
```

```python
#we loop through every page and add each result to our list
#we make sure it is a list and use time to stay under limit
full_movie_list=[]
page=1
count=0
while len(full_movie_list)<20000 and count<2000 and page<1000:
    count+=1
    r=get_popular_movies_TMDB(api_key, page)
    if type(r)==list:
        
        full_movie_list+=r
        page+=1
        
        if page%10==0:
            print(page)
    
    
    time.sleep(.25)
```

```python
# we pull out the movie id which we will use in other part of website
full_movie_id=[]
for x in full_movie_list:
    full_movie_id.append(x['id'])  
```

```python
#check the set to see if its all different
len(set(full_movie_id))
```

```python
#we take the movie id's and use each one in the movie url
#we loop thru every page in the movie data base
#using a while loop with some safety measures to ending
#we ask if it is a dict and the connection in good
#we take the id of each movie into a list for us to use later
#we have it all on a timer so we dont overdo the api pings
#count function so we dont go on forever, we also print the counts every 50 and the length of the set 
#so we can see what is coming out
movies=[]
count=0
ids=[]
while len(movies)< len(full_movie_id) and count< 2*len(full_movie_id):
    movie_url= 'https://api.themoviedb.org/3/movie/' +str(full_movie_id[len(movies)]) + '?api_key={}&language=en-US'.format(api_key)
    res= requests.get(movie_url)
    if type(res.json())==dict and res.status_code==200:
        ids.append(res.json()['id'])
        movies.append(res.json())
    time.sleep(.25)
    count+=1
    if count%50==0:
        print(count)
        print(len(set(ids)))
```
