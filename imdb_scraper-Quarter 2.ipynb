```python
def imdb_scrape(url):
    html_page = requests.get(url)
    soup = BeautifulSoup(html_page.content, 'html.parser')
    rating_info={}
    rating_stuff= soup.findAll(class_="imdbRating")
    try:
        rating_info['imdb_rating']=float(str(rating_stuff).split("ratingValue")[-1].split('>')[1].split('<')[0])
    except:
        pass
    try:
        rating_info['imdb_rating_count']=int(str(rating_stuff).split('ratingCount')[-1].split('>')[1].split('<')[0].replace(',',''))
    except:
        pass
    try:
        rating_info['imdb_critic_reviews']=int(str(rating_stuff).split('reviewCount')[-1].split('>')[1].split('critic')[0].replace(',',''))
    except:
        pass
    try:
        rating_info['imdb_user_reviews']=int(str(rating_stuff).split('reviewCount')[-2].split('>')[1].split('user')[0].replace(',',''))
    except:
        pass
    try:
        meta=soup.findAll(class_="titleReviewBarItem")
    except:
        pass
    try:
        rating_info['imdb_meta_score']=int(str(meta).split('metacritic')[1].split('span>')[1].split('<')[0])
    except:
        pass
    try:
        rating_info["imdb_popularity"]=int(str(meta).split('Popularity')[-1].split('subText')[1].split('\n')[1].strip().replace(',',''))
    except:
        pass
    try:
        people=soup.findAll(class_='credit_summary_item')
    except:
        pass
    try:
        rating_info['imdb_writers']=people[1].getText().split('\n')[2].split('|')[0].replace('(screenplay)','').strip().split(',')
    except:
        pass
    try:
        rating_info["imdb_director"]=people[0].getText().split('\n')[2].split('|')[0].replace('(screenplay)','').strip().split(',')[0]
    except:
        pass
    try:
        rating_info['imdb_stars']= people[2].getText().split('\n')[2].split('|')[0].replace('(screenplay)','').strip().split(',')
    except:
        pass
    return rating_info
    
```

```python
with open('/Users/denisdunn/flatiron/nyc-mhtn-ds-031119-lectures/project1/imdb_list.txt') as f:
    lines = f.read().splitlines()
```

```python
clean_lines=[]
for line in lines:
    if line!="None":
        clean_lines.append(line)
```

```python
len(clean_lines)
```

```python
import time
import json
```

```python
import random
```

```python
quarter_list=[]
for i in range(0, len(clean_lines)):
    if i%4==2:
        quarter_list.append(clean_lines[i])
```

```python
imdb_dicts=[]
rating_counts=[]
count=0

while len(imdb_dicts)<len(quarter_list) and count<2*len(quarter_list):
    
    start=time.time()
    ran=random.random()
    count+=1
    url= "https://www.imdb.com/title/"  + quarter_list[len(imdb_dicts)]
    imdb_data=imdb_scrape(url)
    imdb_data['imdb_key']= quarter_list[len(imdb_dicts)]
    imdb_dicts.append(imdb_data)
    try:
        rating_counts.append(imdb_data['imdb_rating_count'])
    except:
        pass
    if count%50==0:
        print(count)
        print(len(set(rating_counts)))
        if count%200==0:
            with open('imdb_quart_2', 'w') as f:
                json.dump(imdb_dicts, f)
                print('saved')
    end=time.time()
    q=max(start+0.5-end + ran , ran/5)
    time.sleep(q)
    
with open('imdb_quart_2', 'w') as f:
    json.dump(imdb_dicts, f)
    print('LIST SAVE DONE')

```
